## The character finds a weapon

Weapons can be found using an instance method, the name of which is the name of the weapon. The name of the weapon method can be anything, but always consists of 3 words: the type of weapon (for example, "axe"), the word "of" and the property of the weapon (for example, "fire").

ch.axe_of_fire(3, 1, 0, 20) # weapon method name can be anything as long as it matches weapon_of_something
Weapon damage is set by parameters passed to the method, which give damage depending on characteristics and additional damage.(always the order: strength, dexterity, intellect, extra damage), the number of parameters are always 4.

In this case, the Axe of fire has parameters 3, 1, 0, 20. This means that the damage from this weapon will be calculated according to the formula

```python
from typing import Any


class Character:
    def __init__(
        self,
        name: str = "Hero",
        strength: int = 10,
        dexterity: int = 10,
        intelligence: int = 10,
    ) -> None:
        self.name = name
        self.strength = strength
        self.dexterity = dexterity
        self.intelligence = intelligence

        self.weapon = "limbs"
        self.damage = self.strength + self.dexterity + self.intelligence

    def limbs(self):
        return self.strength + self.dexterity + self.intelligence

    def base_weapon(self, s, d, i, base, name=None):
        dmg = self.strength * s + self.dexterity * d + self.intelligence * i + base
        if name and dmg > self.damage:
            self.damage = dmg
            self.weapon = setattr(self, name, self.base_weapon)
        return dmg

    def character_info(self) -> str:
        """Return a multiline string representing the current characteristics of the hero:

        Kroker
        str 15
        dex 10
        int 7
        limbs 32 dmg
        """
        return (
            f"{self.name}\nstr {self.strength}\ndex {self.dexterity}\n"
            + f"int {self.intelligence}\n{self.weapon} {self.damage} dmg"
        )

    def event_log(self) -> str:
        """Return a multiline string representing the events that occured in chronological order"""

    def __getattr__(self, name: str) -> Any:
        nm = name.split("_")
        if len(nm) == 3 and nm[1] == "of":
            self.weapon = name.capitalize().replace("_", " ")
            setattr(self, name, self.base_weapon)
            return self.__dict__[name]


ch = Character(name="Kroker", strength=15, intelligence=7)
print(ch.character_info())

ch.axe_of_fire(3, 1, 0, 20)
print(ch.character_info())
```

```python
from functools import wraps
from typing import Any


def decorate_functions(function_decorator):
    def decorator(cls):
        for name, obj in vars(cls).items():
            if callable(obj):
                try:
                    obj = obj.__func__
                except AttributeError:
                    pass
                setattr(cls, name, function_decorator(obj))
        return cls

    return decorator


def print_on_call(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f"{func.__name__} called")
        try:
            res = func(*args, **kwargs)
        finally:
            print(f"{func.__name__} finnished")
        return res

    return wrapper


@decorate_functions(print_on_call)
class Foo:
    def f1(self):
        print(1)

    def f2(self):
        print(2)


c = Foo()

c.f1()
c.f2()
c.f3()
c.fsdklj()
```
